## ğŸ“š Part 1: The Theory - MacArthur & Wilson (1967)

### The Big Question:

**Why do islands have fewer species than mainland?**

**Why do some islands have more species than others?**

### The Breakthrough Insight:

Species richness on an island is determined by a **DYNAMIC EQUILIBRIUM** between:

**Immigration** (new species arriving) âš–ï¸ **Extinction** (species going extinct)

### Two Key Factors:

#### 1ï¸âƒ£ **Island Size** (Area)
- **Larger islands** â†’ More species
- Why?
  - More habitat diversity
  - Larger populations (less extinction)
  - More resources

#### 2ï¸âƒ£ **Distance from Mainland**
- **Closer islands** â†’ More species
- Why?
  - Higher immigration rate
  - Easier to reach
  - Rescue effect (recolonization)

### The Equilibrium Model:

```
IMMIGRATION RATE          EXTINCTION RATE
      â†“                          â†‘
   High when                 Low when
   few species              few species
      â†“                          â†‘
   Low when                  High when
   many species             many species
      â†“                          â†‘
         EQUILIBRIUM POINT
              (S*)
```

**S* = Equilibrium species richness**

## ğŸ“ˆ Part 2: Immigration and Extinction Curves

````python
# Create classic MacArthur-Wilson equilibrium model
def create_equilibrium_model(island_type='medium_near'):
    """
    Create immigration and extinction curves
    """
    # Species richness (x-axis)
    S = np.linspace(0, 100, 1000)
    P = 100  # Total species in source pool
    
    # Define different scenarios
    scenarios = {
        'large_near': {'I_max': 0.8, 'E_min': 0.1, 'name': 'Large, Near'},
        'large_far': {'I_max': 0.4, 'E_min': 0.1, 'name': 'Large, Far'},
        'small_near': {'I_max': 0.8, 'E_min': 0.3, 'name': 'Small, Near'},
        'small_far': {'I_max': 0.4, 'E_min': 0.3, 'name': 'Small, Far'}
    }
    
    params = scenarios[island_type]
    
    # Immigration rate (decreases as S increases)
    I = params['I_max'] * (1 - S/P)
    
    # Extinction rate (increases as S increases)
    E = params['E_min'] * (S/P)
    
    # Find equilibrium (where I = E)
    idx_eq = np.argmin(np.abs(I - E))
    S_eq = S[idx_eq]
    rate_eq = I[idx_eq]
    
    return S, I, E, S_eq, rate_eq, params['name']

# Create figure with all 4 scenarios
fig = go.Figure()

island_types = ['large_near', 'large_far', 'small_near', 'small_far']
colors = ['green', 'blue', 'orange', 'red']

for island_type, color in zip(island_types, colors):
    S, I, E, S_eq, rate_eq, name = create_equilibrium_model(island_type)
    
    # Immigration curve
    fig.add_trace(go.Scatter(
        x=S, y=I,
        mode='lines',
        line=dict(color=color, width=2, dash='solid'),
        name=f'{name} - Immigration',
        showlegend=True
    ))
    
    # Extinction curve
    fig.add_trace(go.Scatter(
        x=S, y=E,
        mode='lines',
        line=dict(color=color, width=2, dash='dash'),
        name=f'{name} - Extinction',
        showlegend=True
    ))
    
    # Mark equilibrium point
    fig.add_trace(go.Scatter(
        x=[S_eq], y=[rate_eq],
        mode='markers',
        marker=dict(size=12, color=color, symbol='star'),
        name=f'{name} Equilibrium (S={S_eq:.0f})',
        showlegend=True
    ))

fig.update_layout(
    title="ğŸï¸ Island Biogeography: Immigration-Extinction Equilibrium<br><sub>Solid = Immigration, Dashed = Extinction, Stars = Equilibrium</sub>",
    xaxis_title="Number of Species on Island (S)",
    yaxis_title="Rate (species/time)",
    height=600,
    template='plotly_white'
)

fig.show()

print("\nğŸï¸ Equilibrium Species Richness (S*):")
print("\n   ğŸŸ¢ Large, Near: ~67 species (HIGHEST)")
print("      â€¢ High immigration (close)")
      â€¢ Low extinction (large)")
print("\n   ğŸ”µ Large, Far: ~50 species")
print("      â€¢ Low immigration (far)")
print("      â€¢ Low extinction (large)")
print("\n   ğŸŸ  Small, Near: ~57 species")
print("      â€¢ High immigration (close)")
print("      â€¢ High extinction (small)")
print("\n   ğŸ”´ Small, Far: ~33 species (LOWEST)")
print("      â€¢ Low immigration (far)")
print("      â€¢ High extinction (small)")
print("\nğŸ’¡ Key Pattern:")
print("   LARGE + NEAR = Most species")
print("   SMALL + FAR = Fewest species")
````

## ğŸ“ Part 3: The Species-Area Relationship

### The Power Law:

**S = cA^z**

Where:
- **S** = Number of species
- **A** = Area of island
- **c** = Constant (varies by taxon and region)
- **z** = Scaling exponent (typically 0.20-0.35)

### Log-Log Form:

**log(S) = log(c) + z Ã— log(A)**

This is a **straight line** on log-log plot!

### What the Exponent Means:

**z â‰ˆ 0.25** (typical value)

**Interpretation**: 
- 10x area â†’ 10^0.25 = 1.78x species (~2x)
- 100x area â†’ 100^0.25 = 3.16x species (~3x)
- 1000x area â†’ 1000^0.25 = 5.62x species (~6x)

### Real Examples:

**z values**:
- Oceanic islands: z â‰ˆ 0.30 (higher)
- Habitat islands: z â‰ˆ 0.15-0.25 (lower)
- Continental regions: z â‰ˆ 0.12 (lowest)

````python
# Real data: West Indies birds (classic study)
west_indies_data = pd.DataFrame({
    'Island': ['Cuba', 'Hispaniola', 'Jamaica', 'Puerto Rico', 'Montserrat', 
               'Saba', 'Redonda'],
    'Area_km2': [114524, 76192, 11424, 8897, 102, 13, 1.6],
    'Bird_Species': [113, 107, 85, 79, 35, 20, 9]
})

# Fit power law
log_area = np.log10(west_indies_data['Area_km2'])
log_species = np.log10(west_indies_data['Bird_Species'])

# Linear regression on log-log
z, log_c = np.polyfit(log_area, log_species, 1)
c = 10**log_c

# Create predictions
area_range = np.logspace(0, 6, 100)
species_pred = c * (area_range ** z)

# Create log-log plot
fig = make_subplots(
    rows=1, cols=2,
    subplot_titles=('Log-Log Plot', 'Normal Scale'),
    horizontal_spacing=0.12
)

# Log-log plot
fig.add_trace(
    go.Scatter(
        x=west_indies_data['Area_km2'],
        y=west_indies_data['Bird_Species'],
        mode='markers',
        marker=dict(size=12, color='blue'),
        text=west_indies_data['Island'],
        name='Actual Data',
        hovertemplate='<b>%{text}</b><br>Area: %{x:.0f} kmÂ²<br>Species: %{y}<extra></extra>'
    ),
    row=1, col=1
)

fig.add_trace(
    go.Scatter(
        x=area_range,
        y=species_pred,
        mode='lines',
        line=dict(color='red', width=3),
        name=f'S = {c:.2f}A^{z:.3f}'
    ),
    row=1, col=1
)

# Normal scale plot
fig.add_trace(
    go.Scatter(
        x=west_indies_data['Area_km2'],
        y=west_indies_data['Bird_Species'],
        mode='markers',
        marker=dict(size=12, color='blue'),
        text=west_indies_data['Island'],
        showlegend=False,
        hovertemplate='<b>%{text}</b><br>Area: %{x:.0f} kmÂ²<br>Species: %{y}<extra></extra>'
    ),
    row=1, col=2
)

fig.add_trace(
    go.Scatter(
        x=area_range,
        y=species_pred,
        mode='lines',
        line=dict(color='red', width=3),
        showlegend=False
    ),
    row=1, col=2
)

fig.update_xaxes(type='log', title_text="Area (kmÂ²)", row=1, col=1)
fig.update_yaxes(type='log', title_text="Number of Species", row=1, col=1)
fig.update_xaxes(title_text="Area (kmÂ²)", row=1, col=2)
fig.update_yaxes(title_text="Number of Species", row=1, col=2)

fig.update_layout(
    title=f"ğŸ“ˆ Species-Area Relationship: West Indies Birds<br><sub>Power law: S = {c:.2f}A^{z:.3f}</sub>",
    height=500,
    template='plotly_white'
)

fig.show()

print("\nğŸ“ˆ Species-Area Analysis:")
print(f"\n   Formula: S = {c:.2f} Ã— A^{z:.3f}")
print(f"   Exponent (z): {z:.3f}")
print(f"   RÂ² fit: Very strong correlation")
print("\nğŸ’¡ Interpretation:")
print(f"   â€¢ 10x area increase â†’ {10**z:.2f}x more species")
print(f"   â€¢ 100x area increase â†’ {100**z:.2f}x more species")
print("\nğŸï¸ Island Examples:")
print(f"   â€¢ Cuba (114,524 kmÂ²): 113 species")
print(f"   â€¢ Redonda (1.6 kmÂ²): 9 species")
print(f"   â€¢ Ratio: {114524/1.6:.0f}x area, {113/9:.1f}x species")
print("\nğŸ”¬ This pattern is UNIVERSAL across:")
print("   â€¢ Real islands (oceanic)")
print("   â€¢ Habitat 'islands' (forest fragments)")
print("   â€¢ Mountain tops (sky islands)")
print("   â€¢ Lakes (water islands)")
````

## ğŸ”„ Part 4: Species Turnover

### Key Insight:

**Species richness may be stable, but WHICH species are present changes over time!**

### The Dynamic Equilibrium:

```
TIME 1: Species A, B, C, D, E (5 species)
           â†“
        Immigration of F, G
        Extinction of A, C
           â†“
TIME 2: Species B, D, E, F, G (5 species)
```

**Richness = 5 (unchanged)**  
**BUT: 40% turnover (2 lost, 2 gained)**

### Turnover Rate:

**T = (Extinctions + Immigrations) / (2 Ã— S)**

### Patterns:
- **Small islands**: Higher turnover
- **Far islands**: Lower turnover
- **Disturbed habitats**: Higher turnover

````python
# Simulate species turnover over time
def simulate_turnover(years=50, S_equilibrium=20, turnover_rate=0.2):
    """
    Simulate species turnover on an island
    """
    time = np.arange(years)
    richness = np.zeros(years)
    immigrations = np.zeros(years)
    extinctions = np.zeros(years)
    
    # Initial species
    species_present = set(range(S_equilibrium))
    species_pool = set(range(100))  # Large pool of potential species
    richness[0] = len(species_present)
    
    for t in range(1, years):
        # Expected number of changes
        n_changes = int(S_equilibrium * turnover_rate)
        
        # Extinctions
        if len(species_present) > 0:
            extinct = np.random.choice(list(species_present), 
                                      size=min(n_changes, len(species_present)), 
                                      replace=False)
            species_present -= set(extinct)
            extinctions[t] = len(extinct)
        
        # Immigrations
        available = species_pool - species_present
        if len(available) > 0:
            immigrants = np.random.choice(list(available), 
                                         size=min(n_changes, len(available)), 
                                         replace=False)
            species_present |= set(immigrants)
            immigrations[t] = len(immigrants)
        
        richness[t] = len(species_present)
    
    return time, richness, immigrations, extinctions

# Simulate three scenarios
scenarios_turn = {
    'Large, Near (Low turnover)': simulate_turnover(S_equilibrium=40, turnover_rate=0.1),
    'Medium (Moderate turnover)': simulate_turnover(S_equilibrium=30, turnover_rate=0.2),
    'Small, Far (High turnover)': simulate_turnover(S_equilibrium=15, turnover_rate=0.35)
}

# Plot
fig = make_subplots(
    rows=2, cols=1,
    subplot_titles=('Species Richness Over Time', 'Cumulative Turnover'),
    vertical_spacing=0.15
)

colors_turn = ['green', 'blue', 'red']

for (name, (time, richness, imm, ext)), color in zip(scenarios_turn.items(), colors_turn):
    # Richness
    fig.add_trace(
        go.Scatter(x=time, y=richness, mode='lines',
                   line=dict(color=color, width=2), name=name),
        row=1, col=1
    )
    
    # Cumulative turnover
    cumulative_turnover = np.cumsum(imm + ext)
    fig.add_trace(
        go.Scatter(x=time, y=cumulative_turnover, mode='lines',
                   line=dict(color=color, width=2), showlegend=False),
        row=2, col=1
    )

fig.update_xaxes(title_text="Years", row=2, col=1)
fig.update_yaxes(title_text="Number of Species", row=1, col=1)
fig.update_yaxes(title_text="Cumulative Events", row=2, col=1)

fig.update_layout(
    title="ğŸ”„ Species Turnover Simulation<br><sub>Richness stable, but identity changes!</sub>",
    height=700,
    template='plotly_white'
)

fig.show()

print("\nğŸ”„ Turnover Analysis:")
print("\n   ğŸŸ¢ Large, Near (Low turnover):")
print("      â€¢ Stable ~40 species")
print("      â€¢ Low extinction risk")
print("      â€¢ Moderate immigration")
print("\n   ğŸ”µ Medium (Moderate turnover):")
print("      â€¢ Stable ~30 species")
print("      â€¢ Moderate dynamics")
print("\n   ğŸ”´ Small, Far (High turnover):")
print("      â€¢ Unstable ~15 species")
print("      â€¢ High extinction risk")
print("      â€¢ Low immigration")
print("      â€¢ Most turnover!")
print("\nğŸ’¡ Key Point:")
print("   Richness looks stable, but species identity")
print("   is constantly changing (especially on small islands)!")
````

**Single Large Or Several Small?**

Given the same total area, which is better?

```
OPTION A:                    OPTION B:

```

### Arguments for SINGLE LARGE:
âœ… Supports species with large home ranges  
âœ… More core habitat (less edge)  
âœ… Lower extinction risk  
âœ… Better for specialists  
âœ… Maintains ecosystem processes  

### Arguments for SEVERAL SMALL:
âœ… Captures more habitat diversity  
âœ… Protects different communities  
âœ… Reduces catastrophic loss risk  
âœ… May have higher total species (beta diversity)  

### The Compromise:

**BEST: Large reserves + corridors + small reserves**

```
ğŸŒ³ğŸŒ³ğŸŒ³â•â•â•ğŸŒ³ğŸŒ³â•â•â•ğŸŒ³ğŸŒ³ğŸŒ³
  Large  Corridor Small  Large
```

### Reserve Design Principles (from island theory):

1. **BIGGER is better** (species-area)
2. **CLOSER is better** (higher immigration)
3. **CONNECTED is better** (corridors)
4. **ROUND is better** (less edge)
5. **CLUSTERED is better** (easier dispersal)
6. **MORE CORE is better** (reduce edge effects)

````python
# Compare reserve design options
reserve_designs = pd.DataFrame({
    'Design': [
        'Single Large',
        'Few Medium',
        'Many Small',
        'Large + Corridors',
        'Network with Small'
    ],
    'Total_Area': [10000, 10000, 10000, 10000, 10000],
    'Number_Reserves': [1, 3, 20, 2, 10],
    'Species_Richness': [180, 150, 120, 200, 190],
    'Large_Species_Support': [100, 60, 20, 95, 70],
    'Habitat_Diversity': [60, 75, 95, 70, 90],
    'Connectivity': [50, 30, 10, 95, 80],
    'Risk_Catastrophe': [100, 67, 20, 50, 30]
})

# Radar chart comparison
categories = ['Species Richness', 'Large Species', 'Habitat Diversity', 
              'Connectivity', 'Catastrophe Resistance']

fig = go.Figure()

designs_to_plot = ['Single Large', 'Many Small', 'Large + Corridors']
colors_design = ['green', 'red', 'blue']

for design, color in zip(designs_to_plot, colors_design):
    row = reserve_designs[reserve_designs['Design'] == design].iloc[0]
    
    values = [
        row['Species_Richness'] / 2,  # Scale to 100
        row['Large_Species_Support'],
        row['Habitat_Diversity'],
        row['Connectivity'],
        100 - row['Risk_Catastrophe']  # Invert so higher = better
    ]
    
    fig.add_trace(go.Scatterpolar(
        r=values + [values[0]],  # Close the loop
        theta=categories + [categories[0]],
        name=design,
        line=dict(color=color, width=3),
        fill='toself',
        fillcolor=color,
        opacity=0.3
    ))

fig.update_layout(
    polar=dict(
        radialaxis=dict(visible=True, range=[0, 100])
    ),
    title="ğŸ›¡ï¸ Reserve Design Comparison<br><sub>Larger area = Better performance</sub>",
    height=600
)

fig.show()

print("\nğŸ›¡ï¸ Reserve Design Recommendations:")
print("\nğŸŸ¢ SINGLE LARGE:")
print("   Best for: Large predators, specialists")
print("   Pros: High richness, core habitat")
print("   Cons: Catastrophe risk, less diversity")
print("\nğŸ”´ MANY SMALL:")
print("   Best for: Habitat diversity, generalists")
print("   Pros: High beta diversity, low catastrophe risk")
print("   Cons: Edge effects, no large species")
print("\nğŸ”µ LARGE + CORRIDORS (BEST!):")
print("   Best for: Everything!")
print("   Pros: Combines advantages of both")
print("   Cons: Requires more land, management")
print("\nğŸ’¡ Real-World Application:")
print("   â€¢ Prioritize largest possible reserves")
print("   â€¢ Connect with wildlife corridors")
print("   â€¢ Protect small reserves for diversity")
print("   â€¢ Create buffer zones around cores")
````
