## ğŸ® Part 3: Interactive Distribution Explorer

### Filter 1: Binomial Distribution - Success/Failure Filter ğŸŒ±

**Ecological Example**: Out of 20 seeds planted, how many will germinate?

**When to use**:
- Fixed number of trials (n)
- Each trial: Success or Failure
- Trials are independent
- Probability stays constant (p)

**Parameters**:
- `n` = number of trials (seeds planted)
- `p` = probability of success (germination rate)

````python
# Interactive Binomial Distribution Explorer
def plot_binomial(n=20, p=0.7):
    """
    Interactive binomial distribution plotter
    """
    # Generate distribution
    x = np.arange(0, n+1)
    pmf = stats.binom.pmf(x, n, p)
    
    # Calculate statistics
    mean = n * p
    variance = n * p * (1-p)
    std = np.sqrt(variance)
    
    # Create figure
    fig = go.Figure()
    
    # Add bars
    fig.add_trace(go.Bar(
        x=x,
        y=pmf,
        marker_color='#4CAF50',
        name='Probability',
        hovertemplate='<b>%{x} seeds germinate</b><br>Probability: %{y:.3f}<extra></extra>'
    ))
    
    # Add mean line
    fig.add_vline(x=mean, line_dash="dash", line_color="red",
                  annotation_text=f"Mean = {mean:.1f}")
    
    # Update layout
    fig.update_layout(
        title=f"ğŸŒ± Binomial Distribution: Seed Germination<br>n={n} seeds, p={p:.2f} germination rate",
        xaxis_title="Number of Seeds Germinated",
        yaxis_title="Probability",
        hovermode='x',
        height=500,
        template='plotly_white'
    )
    
    # Display statistics
    stats_text = f"""
    <div style='background: #e8f5e9; padding: 15px; border-radius: 10px; margin-top: 10px;'>
        <h4>ğŸ“Š Statistics:</h4>
        <ul>
            <li><strong>Mean (Î¼)</strong> = nÃ—p = {n}Ã—{p} = <strong>{mean:.2f}</strong> seeds</li>
            <li><strong>Variance (ÏƒÂ²)</strong> = nÃ—pÃ—(1-p) = <strong>{variance:.2f}</strong></li>
            <li><strong>Standard Deviation (Ïƒ)</strong> = <strong>{std:.2f}</strong> seeds</li>
        </ul>
        <p><strong>Interpretation:</strong> On average, {mean:.1f} seeds will germinate out of {n}, 
        with typical variation of Â±{std:.1f} seeds.</p>
    </div>
    """
    
    fig.show()
    display(HTML(stats_text))

# Create interactive widget
interact(plot_binomial,
         n=widgets.IntSlider(min=5, max=50, step=1, value=20, description='Seeds (n):'),
         p=widgets.FloatSlider(min=0.01, max=0.99, step=0.01, value=0.7, description='Success (p):'));
````

### Filter 2: Poisson Distribution - Rare Events Filter ğŸ¦‹

**Ecological Example**: How many new butterfly species colonize an island per year?

**When to use**:
- Counting **rare events** in fixed interval (time/space)
- Events occur independently
- Average rate is constant

**Parameter**:
- `Î»` (lambda) = average rate of events

````python
# Interactive Poisson Distribution Explorer
def plot_poisson(lambda_val=3.0):
    """
    Interactive Poisson distribution plotter
    """
    # Generate distribution
    x = np.arange(0, int(lambda_val * 3 + 10))
    pmf = stats.poisson.pmf(x, lambda_val)
    
    # Statistics
    mean = lambda_val
    variance = lambda_val
    std = np.sqrt(variance)
    
    # Create figure
    fig = go.Figure()
    
    fig.add_trace(go.Bar(
        x=x,
        y=pmf,
        marker_color='#2196F3',
        name='Probability',
        hovertemplate='<b>%{x} colonization events</b><br>Probability: %{y:.3f}<extra></extra>'
    ))
    
    fig.add_vline(x=mean, line_dash="dash", line_color="red",
                  annotation_text=f"Mean = Î» = {mean:.1f}")
    
    fig.update_layout(
        title=f"ğŸ¦‹ Poisson Distribution: Species Colonization<br>Î» = {lambda_val:.1f} species per year",
        xaxis_title="Number of Colonization Events",
        yaxis_title="Probability",
        height=500,
        template='plotly_white'
    )
    
    stats_text = f"""
    <div style='background: #e3f2fd; padding: 15px; border-radius: 10px; margin-top: 10px;'>
        <h4>ğŸ“Š Statistics:</h4>
        <ul>
            <li><strong>Mean (Î¼)</strong> = Î» = <strong>{mean:.2f}</strong> events</li>
            <li><strong>Variance (ÏƒÂ²)</strong> = Î» = <strong>{variance:.2f}</strong></li>
            <li><strong>Standard Deviation (Ïƒ)</strong> = âˆšÎ» = <strong>{std:.2f}</strong> events</li>
        </ul>
        <p><strong>Key Property:</strong> For Poisson, mean = variance = Î»<br>
        <strong>Interpretation:</strong> On average, {mean:.1f} new species colonize per year.</p>
    </div>
    """
    
    fig.show()
    display(HTML(stats_text))

interact(plot_poisson,
         lambda_val=widgets.FloatSlider(min=0.5, max=15, step=0.5, value=3.0, 
                                       description='Rate (Î»):'));
````

<details>

    <li>As Î» increases, the distribution becomes more <strong>symmetric</strong> and normal-looking</li>

### Filter 3: Normal Distribution - Natural Variation Filter ğŸ“

**Ecological Example**: Heights of adult plants in a population

**When to use**:
- Continuous measurements
- Many small factors contributing to variation
- Symmetric variation around mean

**Parameters**:
- `Î¼` (mu) = mean (center)
- `Ïƒ` (sigma) = standard deviation (spread)

````python
# Interactive Normal Distribution Explorer
def plot_normal(mu=50, sigma=10):
    """
    Interactive normal distribution plotter
    """
    # Generate distribution
    x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
    pdf = stats.norm.pdf(x, mu, sigma)
    
    # Create figure
    fig = go.Figure()
    
    # Add curve
    fig.add_trace(go.Scatter(
        x=x,
        y=pdf,
        mode='lines',
        fill='tozeroy',
        line=dict(color='#FF9800', width=3),
        name='Probability Density',
        hovertemplate='Height: %{x:.1f} cm<br>Density: %{y:.4f}<extra></extra>'
    ))
    
    # Add mean and SD lines
    fig.add_vline(x=mu, line_dash="solid", line_color="red", line_width=2,
                  annotation_text=f"Î¼ = {mu}")
    fig.add_vline(x=mu-sigma, line_dash="dash", line_color="blue",
                  annotation_text=f"Î¼-Ïƒ")
    fig.add_vline(x=mu+sigma, line_dash="dash", line_color="blue",
                  annotation_text=f"Î¼+Ïƒ")
    
    # 68-95-99.7 rule regions
    x_68 = x[(x >= mu-sigma) & (x <= mu+sigma)]
    pdf_68 = stats.norm.pdf(x_68, mu, sigma)
    
    fig.add_trace(go.Scatter(
        x=x_68,
        y=pdf_68,
        fill='tozeroy',
        mode='none',
        fillcolor='rgba(76, 175, 80, 0.3)',
        name='68% of data',
        showlegend=True
    ))
    
    fig.update_layout(
        title=f"ğŸ“ Normal Distribution: Plant Heights<br>Î¼ = {mu} cm, Ïƒ = {sigma} cm",
        xaxis_title="Plant Height (cm)",
        yaxis_title="Probability Density",
        height=500,
        template='plotly_white'
    )
    
    stats_text = f"""
    <div style='background: #fff3e0; padding: 15px; border-radius: 10px; margin-top: 10px;'>
        <h4>ğŸ“Š Statistics & The 68-95-99.7 Rule:</h4>
        <ul>
            <li><strong>Mean (Î¼)</strong> = <strong>{mu:.1f}</strong> cm (center of distribution)</li>
            <li><strong>Standard Deviation (Ïƒ)</strong> = <strong>{sigma:.1f}</strong> cm (measure of spread)</li>
        </ul>
        <h4>ğŸ¯ Empirical Rule:</h4>
        <ul>
            <li><strong>68%</strong> of plants are between {mu-sigma:.1f} and {mu+sigma:.1f} cm (Î¼ Â± 1Ïƒ)</li>
            <li><strong>95%</strong> are between {mu-2*sigma:.1f} and {mu+2*sigma:.1f} cm (Î¼ Â± 2Ïƒ)</li>
            <li><strong>99.7%</strong> are between {mu-3*sigma:.1f} and {mu+3*sigma:.1f} cm (Î¼ Â± 3Ïƒ)</li>
        </ul>
        <p><strong>Shape:</strong> Perfectly symmetric bell curve - the most common pattern in nature!</p>
    </div>
    """
    
    fig.show()
    display(HTML(stats_text))

interact(plot_normal,
         mu=widgets.IntSlider(min=20, max=100, step=5, value=50, description='Mean (Î¼):'),
         sigma=widgets.IntSlider(min=1, max=30, step=1, value=10, description='SD (Ïƒ):'));
````

### ğŸ” Pattern Discovery Questions

**Explore**:
1. What happens when Ïƒ is very small (like 2)?
2. What happens when Ïƒ is large (like 25)?
3. Does changing Î¼ change the shape, or just shift it?

<details>
<summary>ğŸ’¡ Click for insights</summary>
<ul>
    <li>Small Ïƒ â†’ <strong>narrow, tall</strong> peak (less variation)</li>
    <li>Large Ïƒ â†’ <strong>wide, short</strong> peak (more variation)</li>
    <li>Changing Î¼ just <strong>shifts</strong> the distribution left/right, doesn't change shape</li>
    <li>The shape is always <strong>symmetric</strong> and bell-shaped</li>
    <li>Mean = Median = Mode (all at the center)</li>

## ğŸ¯ Part 4: Compare Multiple Distributions

Let's compare different distributions side-by-side to see how they differ.

````python
# Generate sample data from different distributions
np.random.seed(42)

# Create figure with subplots
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
fig.suptitle('ğŸ©º The 6 Most Common Distributions in Ecology', fontsize=16, fontweight='bold')

# 1. Binomial
ax = axes[0, 0]
x_binom = np.arange(0, 21)
y_binom = stats.binom.pmf(x_binom, 20, 0.7)
ax.bar(x_binom, y_binom, color='#4CAF50', alpha=0.7)
ax.set_title('ğŸŒ± Binomial\n(Seed Germination)', fontweight='bold')
ax.set_xlabel('Number of successes')
ax.set_ylabel('Probability')
ax.axvline(14, color='red', linestyle='--', label='Mean')
ax.legend()

# 2. Poisson
ax = axes[0, 1]
x_pois = np.arange(0, 15)
y_pois = stats.poisson.pmf(x_pois, 3)
ax.bar(x_pois, y_pois, color='#2196F3', alpha=0.7)
ax.set_title('ğŸ¦‹ Poisson\n(Rare Colonization)', fontweight='bold')
ax.set_xlabel('Number of events')
ax.axvline(3, color='red', linestyle='--')

# 3. Normal
ax = axes[0, 2]
x_norm = np.linspace(20, 80, 1000)
y_norm = stats.norm.pdf(x_norm, 50, 10)
ax.plot(x_norm, y_norm, color='#FF9800', linewidth=2)
ax.fill_between(x_norm, y_norm, alpha=0.3, color='#FF9800')
ax.set_title('ğŸ“ Normal\n(Plant Heights)', fontweight='bold')
ax.set_xlabel('Measurement')
ax.axvline(50, color='red', linestyle='--')

# 4. Exponential
ax = axes[1, 0]
x_exp = np.linspace(0, 10, 1000)
y_exp = stats.expon.pdf(x_exp, scale=2)
ax.plot(x_exp, y_exp, color='#9C27B0', linewidth=2)
ax.fill_between(x_exp, y_exp, alpha=0.3, color='#9C27B0')
ax.set_title('â±ï¸ Exponential\n(Time Between Events)', fontweight='bold')
ax.set_xlabel('Waiting time')

# 5. Uniform
ax = axes[1, 1]
x_unif = np.linspace(0, 10, 1000)
y_unif = stats.uniform.pdf(x_unif, 0, 10)
ax.plot(x_unif, y_unif, color='#00BCD4', linewidth=2)
ax.fill_between(x_unif, y_unif, alpha=0.3, color='#00BCD4')
ax.set_title('ğŸ² Uniform\n(Random Position)', fontweight='bold')
ax.set_xlabel('Position')

# 6. Log-Normal
ax = axes[1, 2]
x_lognorm = np.linspace(0.01, 50, 1000)
y_lognorm = stats.lognorm.pdf(x_lognorm, s=0.8, scale=np.exp(2))
ax.plot(x_lognorm, y_lognorm, color='#795548', linewidth=2)
ax.fill_between(x_lognorm, y_lognorm, alpha=0.3, color='#795548')
ax.set_title('ğŸ“Š Log-Normal\n(Population Sizes)', fontweight='bold')
ax.set_xlabel('Size')

plt.tight_layout()
plt.show()

print("\nâœ… Notice how each distribution has a unique SHAPE!")
print("ğŸ©º Each shape tells a different ecological story.")
````

## ğŸ“š Part 5: Quick Reference Guide

### When to Use Which Distribution?

| Pattern You See | Distribution | Example |
|----------------|--------------|----------|
| Fixed trials, yes/no | **Binomial** | Seed germination |
| Rare events, fixed interval | **Poisson** | Species per year |
| Continuous, symmetric | **Normal** | Plant heights |
| All outcomes equal | **Uniform** | Random position |
| Waiting time | **Exponential** | Time between births |
| Trials until success | **Geometric** | Seeds until sprout |
| Clumped/aggregated | **Neg. Binomial** | Parasites per host |
| Multiplicative, skewed | **Log-Normal** | Population sizes |
