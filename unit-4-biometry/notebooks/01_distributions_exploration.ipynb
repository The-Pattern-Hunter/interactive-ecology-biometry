## ğŸ® Part 2: Interactive Distribution Explorer

### Filter 1: Binomial Distribution ğŸŒ±

**Ecological Example**: Out of 20 seeds planted, how many will germinate?

**When to use**:
- Fixed number of trials (n)
- Each trial: Success or Failure
- Independent trials
- Constant probability (p)

**Use the sliders below to explore!**

````python
# Interactive Binomial Distribution with Plotly Sliders
def create_binomial_plot():
    # Create figure
    fig = go.Figure()
    
    # Generate data for different parameter values
    n_values = list(range(5, 51, 5))
    p_values = [round(p, 2) for p in np.arange(0.1, 1.0, 0.1)]
    
    # Create traces for all combinations (initially invisible)
    for n in n_values:
        for p in p_values:
            x = np.arange(0, n+1)
            pmf = stats.binom.pmf(x, n, p)
            mean = n * p
            std = np.sqrt(n * p * (1-p))
            
            fig.add_trace(go.Bar(
                x=x,
                y=pmf,
                name=f'n={n}, p={p}',
                marker_color='#4CAF50',
                visible=False,
                hovertemplate='<b>%{x} seeds germinate</b><br>Probability: %{y:.3f}<extra></extra>'
            ))
    
    # Make first trace visible
    fig.data[0].visible = True
    
    # Create sliders
    steps_n = []
    for i, n in enumerate(n_values):
        step = dict(
            method="update",
            args=[{"visible": [False] * len(fig.data)},
                  {"title": f"ğŸŒ± Binomial: n={n} seeds, p=0.5"}],
            label=str(n)
        )
        # Make traces visible for this n value with p=0.5
        idx = i * len(p_values) + 4  # p=0.5 is at index 4
        step["args"][0]["visible"][idx] = True
        steps_n.append(step)
    
    steps_p = []
    for j, p in enumerate(p_values):
        step = dict(
            method="update",
            args=[{"visible": [False] * len(fig.data)},
                  {"title": f"ğŸŒ± Binomial: n=20, p={p:.1f}"}],
            label=f"{p:.1f}"
        )
        # Make traces visible for n=20 (index 3) with this p
        idx = 3 * len(p_values) + j
        step["args"][0]["visible"][idx] = True
        steps_p.append(step)
    
    sliders = [
        dict(
            active=3,
            yanchor="top",
            y=1.15,
            xanchor="left",
            currentvalue=dict(prefix="Number of seeds (n): ", visible=True),
            pad=dict(b=10, t=50),
            len=0.9,
            x=0.1,
            steps=steps_n
        ),
        dict(
            active=4,
            yanchor="top",
            y=1.05,
            xanchor="left",
            currentvalue=dict(prefix="Success probability (p): ", visible=True),
            pad=dict(b=10, t=50),
            len=0.9,
            x=0.1,
            steps=steps_p
        )
    ]
    
    fig.update_layout(
        sliders=sliders,
        title="ğŸŒ± Binomial Distribution: Seed Germination<br>n=20 seeds, p=0.5",
        xaxis_title="Number of Seeds Germinated",
        yaxis_title="Probability",
        height=600,
        template='plotly_white',
        showlegend=False
    )
    
    return fig

# Display the plot
fig = create_binomial_plot()
fig.show()

print("\nğŸ“Š Use the sliders above to change n (seeds) and p (success rate)!")
print("ğŸ” Watch how the shape changes!")
````

### ğŸ” Pattern Discovery Questions

**Explore with the sliders above**:

1. Set p = 0.5 â†’ What shape do you see?
2. Set p = 0.1 (low success) â†’ How does the shape change?
3. Increase n to 50 â†’ Does it look smoother?

### Filter 2: Poisson Distribution ğŸ¦‹

**Ecological Example**: How many butterfly species colonize an island per year?

**When to use**: Counting rare events in fixed intervals

**Parameter**: Î» (lambda) = average rate

````python
# Interactive Poisson Distribution
def create_poisson_plot():
    fig = go.Figure()
    
    lambda_values = [0.5, 1, 2, 3, 4, 5, 7, 10, 15]
    
    for lam in lambda_values:
        x = np.arange(0, int(lam * 3 + 15))
        pmf = stats.poisson.pmf(x, lam)
        
        fig.add_trace(go.Bar(
            x=x,
            y=pmf,
            name=f'Î»={lam}',
            marker_color='#2196F3',
            visible=False,
            hovertemplate='<b>%{x} events</b><br>Probability: %{y:.3f}<extra></extra>'
        ))
    
    fig.data[3].visible = True  # Show Î»=3 initially
    
    steps = []
    for i, lam in enumerate(lambda_values):
        step = dict(
            method="update",
            args=[{"visible": [False] * len(fig.data)},
                  {"title": f"ğŸ¦‹ Poisson Distribution: Î»={lam} colonizations/year"}],
            label=str(lam)
        )
        step["args"][0]["visible"][i] = True
        steps.append(step)
    
    sliders = [dict(
        active=3,
        currentvalue=dict(prefix="Rate (Î»): "),
        pad=dict(t=50),
        steps=steps
    )]
    
    fig.update_layout(
        sliders=sliders,
        title="ğŸ¦‹ Poisson Distribution: Species Colonization (Î»=3)",
        xaxis_title="Number of Colonization Events",
        yaxis_title="Probability",
        height=600,
        template='plotly_white',
        showlegend=False
    )
    
    return fig

fig = create_poisson_plot()
fig.show()

print("\nğŸ“Š Slide to change Î» (average rate of events)")
print("ğŸ’¡ Notice: As Î» increases, the distribution becomes more symmetric!")
````

### ğŸ” Pattern Discovery

**Try different Î» values**:
- Î» = 0.5 â†’ Very rare events
- Î» = 3 â†’ Moderate frequency
- Î» = 15 â†’ Looks almost Normal!

**Key insight**: Poisson mean = variance = Î»

### Filter 3: Normal Distribution ğŸ“

**Ecological Example**: Plant heights in a population

**When to use**: Continuous measurements with symmetric variation

**Parameters**: Î¼ (mean), Ïƒ (standard deviation)

````python
# Interactive Normal Distribution
def create_normal_plot():
    fig = go.Figure()
    
    mu_values = [30, 40, 50, 60, 70]
    sigma_values = [5, 10, 15, 20, 25]
    
    for mu in mu_values:
        for sigma in sigma_values:
            x = np.linspace(mu - 4*sigma, mu + 4*sigma, 500)
            pdf = stats.norm.pdf(x, mu, sigma)
            
            fig.add_trace(go.Scatter(
                x=x,
                y=pdf,
                mode='lines',
                fill='tozeroy',
                line=dict(color='#FF9800', width=3),
                name=f'Î¼={mu}, Ïƒ={sigma}',
                visible=False,
                hovertemplate='Height: %{x:.1f}cm<br>Density: %{y:.4f}<extra></extra>'
            ))
    
    # Show Î¼=50, Ïƒ=10 initially (index: 2*5 + 1 = 11)
    fig.data[11].visible = True
    
    steps_mu = []
    for i, mu in enumerate(mu_values):
        step = dict(
            method="update",
            args=[{"visible": [False] * len(fig.data)},
                  {"title": f"ğŸ“ Normal Distribution: Î¼={mu}cm, Ïƒ=10cm"}],
            label=str(mu)
        )
        idx = i * len(sigma_values) + 1  # Ïƒ=10 is at index 1
        step["args"][0]["visible"][idx] = True
        steps_mu.append(step)
    
    steps_sigma = []
    for j, sigma in enumerate(sigma_values):
        step = dict(
            method="update",
            args=[{"visible": [False] * len(fig.data)},
                  {"title": f"ğŸ“ Normal Distribution: Î¼=50cm, Ïƒ={sigma}cm"}],
            label=str(sigma)
        )
        idx = 2 * len(sigma_values) + j  # Î¼=50 is at i=2
        step["args"][0]["visible"][idx] = True
        steps_sigma.append(step)
    
    sliders = [
        dict(
            active=2,
            y=1.15,
            currentvalue=dict(prefix="Mean (Î¼): "),
            pad=dict(t=50),
            steps=steps_mu
        ),
        dict(
            active=1,
            y=1.05,
            currentvalue=dict(prefix="Std Dev (Ïƒ): "),
            pad=dict(t=50),
            steps=steps_sigma
        )
    ]
    
    fig.update_layout(
        sliders=sliders,
        title="ğŸ“ Normal Distribution: Plant Heights (Î¼=50cm, Ïƒ=10cm)",
        xaxis_title="Plant Height (cm)",
        yaxis_title="Probability Density",
        height=600,
        template='plotly_white',
        showlegend=False
    )
    
    return fig

fig = create_normal_plot()
fig.show()

print("\nğŸ“Š Adjust Î¼ (center) and Ïƒ (spread)")
print("ğŸ”” The bell curve - nature's most common pattern!")
````

### ğŸ¯ The 68-95-99.7 Rule

For Normal distribution:
- **68%** of data within Î¼ Â± 1Ïƒ
- **95%** within Î¼ Â± 2Ïƒ
- **99.7%** within Î¼ Â± 3Ïƒ

## ğŸ“Š Part 3: Compare All 6 Distributions

````python
# Create comparison plot
fig = make_subplots(
    rows=2, cols=3,
    subplot_titles=(
        'ğŸŒ± Binomial (n=20, p=0.7)',
        'ğŸ¦‹ Poisson (Î»=3)',
        'ğŸ“ Normal (Î¼=50, Ïƒ=10)',
        'â±ï¸ Exponential (Î»=0.5)',
        'ğŸ² Uniform (a=0, b=10)',
        'ğŸ“Š Log-Normal (Î¼=2, Ïƒ=0.8)'
    ),
    vertical_spacing=0.12,
    horizontal_spacing=0.1
)

# 1. Binomial
x_bin = np.arange(0, 21)
y_bin = stats.binom.pmf(x_bin, 20, 0.7)
fig.add_trace(go.Bar(x=x_bin, y=y_bin, marker_color='#4CAF50', showlegend=False), row=1, col=1)

# 2. Poisson
x_pois = np.arange(0, 15)
y_pois = stats.poisson.pmf(x_pois, 3)
fig.add_trace(go.Bar(x=x_pois, y=y_pois, marker_color='#2196F3', showlegend=False), row=1, col=2)

# 3. Normal
x_norm = np.linspace(20, 80, 300)
y_norm = stats.norm.pdf(x_norm, 50, 10)
fig.add_trace(go.Scatter(x=x_norm, y=y_norm, fill='tozeroy', line=dict(color='#FF9800'), showlegend=False), row=1, col=3)

# 4. Exponential
x_exp = np.linspace(0, 10, 300)
y_exp = stats.expon.pdf(x_exp, scale=2)
fig.add_trace(go.Scatter(x=x_exp, y=y_exp, fill='tozeroy', line=dict(color='#9C27B0'), showlegend=False), row=2, col=1)

# 5. Uniform
x_unif = np.linspace(0, 10, 300)
y_unif = stats.uniform.pdf(x_unif, 0, 10)
fig.add_trace(go.Scatter(x=x_unif, y=y_unif, fill='tozeroy', line=dict(color='#00BCD4'), showlegend=False), row=2, col=2)

# 6. Log-Normal
x_log = np.linspace(0.1, 30, 300)
y_log = stats.lognorm.pdf(x_log, s=0.8, scale=np.exp(2))
fig.add_trace(go.Scatter(x=x_log, y=y_log, fill='tozeroy', line=dict(color='#795548'), showlegend=False), row=2, col=3)

fig.update_layout(
    title_text="ğŸ©º The 6 Most Common Distributions in Ecology",
    height=700,
    showlegend=False,
    template='plotly_white'
)

fig.show()

print("\nâœ¨ Each distribution has a unique SHAPE!")
print("ğŸ©º Each shape tells a different ecological story.")
````
